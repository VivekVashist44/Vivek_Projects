```python
dataset=pd.read_csv('spam_dataset.csv')
print(dataset.shape)

x=dataset.iloc[:,:-1].values
y=dataset.iloc[:,57:].values
print(x.shape)
print(y.shape)

df1=pd.DataFrame(x,columns=['word_freq_make','word_freq_address','word_freq_all','word_freq_3d','word_freq_our','word_freq_over','word_freq_remove','word_freq_internet','word_freq_order','word_freq_mail',
'word_freq_receive','word_freq_will','word_freq_people','word_freq_report','word_freq_addresses','word_freq_free','word_freq_business','word_freq_email','word_freq_you','word_freq_credit',       
'word_freq_your','word_freq_font','word_freq_000','word_freq_money','word_freq_hp','word_freq_hpl','word_freq_george','word_freq_650','word_freq_lab','word_freq_labs','word_freq_telnet',       
'word_freq_857','word_freq_data','word_freq_415','word_freq_85','word_freq_technology','word_freq_1999','word_freq_parts','word_freq_pm','word_freq_direct','word_freq_cs','word_freq_meeting',      
'word_freq_original','word_freq_project','word_freq_re','word_freq_edu','word_freq_table','word_freq_conference','char_freq_;','char_freq_(','char_freq_[','char_freq_!','char_freq_$','char_freq_#','capital_run_length_average','capital_run_length_longest','capital_run_length_total'])
df2=pd.DataFrame(y,columns=["Target"])
df=pd.concat([df1,df2],axis=1)
df

df.describe().T
df.info()

df['Target'].value_counts()
df.isna().sum()

col=['word_freq_make','word_freq_address','word_freq_all','word_freq_3d','word_freq_our','word_freq_over','word_freq_remove','word_freq_internet','word_freq_order','word_freq_mail',
'word_freq_receive','word_freq_will','word_freq_people','word_freq_report','word_freq_addresses','word_freq_free','word_freq_business','word_freq_email','word_freq_you','word_freq_credit',       
'word_freq_your','word_freq_font','word_freq_000','word_freq_money','word_freq_hp','word_freq_hpl','word_freq_george','word_freq_650','word_freq_lab','word_freq_labs','word_freq_telnet',       
'word_freq_857','word_freq_data','word_freq_415','word_freq_85','word_freq_technology','word_freq_1999','word_freq_parts','word_freq_pm','word_freq_direct','word_freq_cs','word_freq_meeting',      
'word_freq_original','word_freq_project','word_freq_re','word_freq_edu','word_freq_table','word_freq_conference','char_freq_;','char_freq_(','char_freq_[','char_freq_!','char_freq_$','char_freq_#','capital_run_length_average','capital_run_length_longest','capital_run_length_total']
cor=df[col].corr()
cor

import seaborn as sns
sns.heatmap(cor,annot=True)

print(y)

from sklearn.impute import SimpleImputer
imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
imputer.fit(x)
x=imputer.transform(x)

print(x)

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0,test_size=0.2)
print(x_train)
print(y_train)

from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)

from sklearn.decomposition import PCA
pca=PCA(n_components=30)
x_train=pca.fit_transform(x_train)
x_test=pca.transform(x_test)
print(x_train.shape)
print(x_test.shape)

from sklearn.linear_model import Perceptron,LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,BaggingClassifier,ExtraTreesClassifier,AdaBoostClassifier,GradientBoostingClassifier,VotingClassifier
cl1=Perceptron()
cl2=LogisticRegression()
cl3=SVC()
cl4=GaussianNB()
cl5=KNeighborsClassifier()
cl6=DecisionTreeClassifier()
cl7=RandomForestClassifier()
cl8=ExtraTreesClassifier()
cl9=AdaBoostClassifier()
cl10=GradientBoostingClassifier()
cl11=VotingClassifier(estimators=[('dt',cl6),('knn',cl5),('svm',cl3)],weights=[1,1,1],voting='hard')
cl12=BaggingClassifier()
cl=[cl1,cl2,cl3,cl4,cl5,cl6,cl7,cl8,cl9,cl10,cl11,cl12]
cl_name=['PER','LR','SVC','GNB','KNN','DTC','RFC','ETC','ABC','GBC','VC','BC']
acc={}
t={}
import time
from sklearn.metrics import accuracy_score

for model,model_name in zip(cl,cl_name):
    st=time.time()
    model.fit(x_train,y_train)
    pred=model.predict(x_test)
    et=time.time()
    acc[model_name]=accuracy_score(y_test,pred)
    t[model_name]=et-st
    
import matplotlib.pyplot as plt
plt.bar(acc.keys(),acc.values())
plt.bar(t.keys(),t.values())

max_acc=max(list(acc.values()))
print(acc)
print(max_acc)
for key,value in acc.items():
    if value==max_acc:
        print("Max accuracy score is of",key,"model.")
        break

from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,cl7.predict(x_test))
print(cm)
sns.heatmap(cm,annot=True,cbar=False)

```
